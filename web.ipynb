```python
# Extracting the data from all pages

titles=[] #List to store name of the product
prices=[] #List to store price of the product
ratings=[] #List to store rating of the product
num_of_rev_ratings = [] #List to store the number of ratings & reviews of the product
specifications = [] # to store the specifications
pgno = []
Discounts = []

for i in range(1,61):
    start_time = time.time()
    URL = 'https://www.flipkart.com/mobile-accessories/power-banks/pr?sid=tyy,4mr,fu6&otracker=categorytree&otracker=nmenu_sub_Electronics_0_Power%20Banks&otracker=nmenu_sub_Electronics_0_Power%20Banks={}'.format(i)
    pagecontent = requests.get(URL).text
    soup = BeautifulSoup(pagecontent)
    #print(URL)
    
    for x  in soup.find_all('div', attrs={'class':"_13oc-S"}):
        title = x.find('a', attrs={'class':'s1Q9rs'})
        if title is None:
            titles.append(np.NaN)
        else:
            titles.append(title.text)

        #_3LWZlK
        rating = x.find('div', attrs={'class':'_3LWZlK'})
        if rating is None:
            ratings.append(np.NaN)
        else:
            ratings.append(rating.text)

        # _2_R_DZ
        num_of_rev_rating = x.find('span', attrs={'class':'_2_R_DZ'})
        if num_of_rev_rating is None:
            num_of_rev_ratings.append(np.NaN)
        else:
            num_of_rev_ratings.append(no_of_rating.text)


        # _30jeq3
        price = x.find('div', attrs={'class':'_30jeq3'})
        if price is None:
            prices.append(np.NaN)
        else:
            prices.append(price.text)
            
        #_3Djpdu
        specification = x.find('div', attrs={'class':'_3Djpdu'})
        if specification is None:
            specifications.append(np.NaN)
        else:
            specifications.append(specification.text)
            
        #_3I9_wc
        Discount = x.find('div', attrs={'class':'_3I9_wc'})
        if Discount is None:
            Discounts.append(np.NaN)
        else:
            Discounts.append(Discount.text)
            
        pgno.append(i)
    
    end_time = time.time()
    print('Page {} completed in {} seconds'. format(i, end_time-start_time))
```

```python
len(Discounts)
```

```python
# build a data frame using the lists

power_bank_df = pd.DataFrame({'PageNo': pgno,'Title':titles,
                              'Price': prices,'specifications':specifications,
                              'Rating': ratings,'Discounts':Discounts,
                              'No_of_R and R': num_of_rev_rating})
```
