## box plot

## scatter plot

## sub plot

```python
plt.figure(dpi = 100, figsize=(15,10))

plt.subplot(221)
sns.heatmap(powerbank_df.corr(), cmap = 'winter', annot=True)
plt.title('Heat map')

plt.subplot(222)
sns.countplot(data = powerbank_df, x = 'Brand')
plt.title('Count plot')

plt.subplot(223)
sns.boxplot(data = powerbank_df, x = 'Brand', y = 'Price')
plt.title('Box plot')

plt.subplot(224)
sns.barplot(data = powerbank_df, x = 'Brand', y = 'Price', ci = None, palette='bright')
plt.title('Bar plot')

plt.tight_layout(h_pad = 2.8, w_pad = 3.5);
```
